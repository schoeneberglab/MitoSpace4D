```python
import numpy as np
import ruptures as rpt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
datafile = "/home/earkfeld/Projects/MitoSpace4D/runs/20260202_3color-embeddings_kinetics-resnet3d/metadata.parquet"
import pandas as pd

df = pd.read_parquet(datafile)

df.head()
```

```python
cell_ids = df['cell_id'].unique().tolist()

cell_derivatives = {}
all_derivatives = []
for cell_id in cell_ids:
    df_cell = df[df['cell_id'] == cell_id]
    # Get the derivatives over consecutive frames
    derivatives = df_cell['embedding'].diff().dropna().values
    cell_derivatives[cell_id] = derivatives
    all_derivatives.append(derivatives)
all_derivatives = np.stack(all_derivatives, axis=0)
```
