```python
import pandas as pd
import numpy as np
import os
```

```python
data_root = '/home/dhruvagarwal/projects/MitoSpace4D/runs/lightning_logs/resnetbilstm_encoded_normal/embeddings/'

print("Loading data...")
emb = np.load(os.path.join(data_root, "embeddings.npy"))         # (34145, 20, 2048)
labels = np.load(os.path.join(data_root, "labels.npy"))     
img_paths = np.load(os.path.join(data_root, "image_file_paths.npy"))
```

```python
df = pd.DataFrame({
    'img_path': img_paths,
    'label': labels,
    'embedding': list(emb)
})
```

```python
# sort by img_path
df = df.sort_values(by='img_path').reset_index(drop=True)
df.head(10)
```

```python
# add pseudo region_id to each row based on img_path: for every label, divide the rows into groups of 100
df['region_id'] = df.groupby('label').cumcount() // 100
df.head(10)
```

```python
# now expand by timepoints
df_expanded = df.explode('embedding').reset_index(drop=True)
df_expanded.head(10)
```

```python
# add timepoint column
df_expanded['timepoint'] = df_expanded.groupby('img_path').cumcount()
df_expanded.head(10)
```

```python
# now add a actual time column based on region_id and timepoint
df_expanded['time'] = df_expanded['region_id'] * 20 + df_expanded['timepoint']
df_expanded.head(10)
```

```python
control_df = df_expanded[df_expanded['label'] == 0].reset_index(drop=True)
len(control_df)
```

```python
drug_labels_dict = {}
label_drug_dict = {}
with open(f"/home/dhruvagarwal/projects/MitoSpace4D/extraction_utils/drugs_to_labels.txt", 'r') as f:
    for line in f:
        folder, drug, label = line.split()
        drug_labels_dict[drug] = int(label)
        label_drug_dict[int(label)] = drug
```

```python
# map labels to drug names
df_expanded['drug'] = df_expanded['label'].map(label_drug_dict)
df_expanded.head(10)
```

```python
df_expanded.to_parquet('expanded_embeddings_with_time.parquet', index=False)
```
